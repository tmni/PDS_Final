```python
import pandas as pd
import numpy as np
```

```python


pop_data = pd.read_csv("Population data/pop_density.csv", header=3)
# pol_data = pd.read_csv("political_data/1976-2018-senate.csv", encoding='utf-8')

race_data_1 = pd.read_csv("Population data/stco-mr2010_al_mo.csv")
race_data_2 = pd.read_csv("Population data/stco-mr2010_mt_wy.csv", encoding='latin-1')

race_data = pd.concat([race_data_1, race_data_2])
# race_data["scaled_sex"] = race_data["SEX"] * race_data["RESPOP"]

# race_data
```
