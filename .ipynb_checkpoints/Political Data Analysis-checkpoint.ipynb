```python
import pandas as pd

pol_data = pd.read_csv("political_data/1976-2016-president.csv", encoding='utf-8')
pol_data
```

```python
tendata = pol_data[pol_data["year"] == 2012]
grouped_by_state= tendata.groupby("state_po").nth(0) #returns highest candidate votes for each state
grouped_by_state
```

```python
sandp = grouped_by_state['party']
sandp = sandp.to_frame()
sandp = sandp.reset_index()
sandp = sandp.set_value(23, 'party', 'democrat')
print(sandp)
sandp = sandp.party.replace(to_replace=['democrat', 'republican'], value=[1, 0])
sandp = sandp.to_frame()
sandp
# sgroup = grouped_by_state.apply(lambda new: new.sort_values(by=['candidatevotes'], ascending=False))
# wow = sgroup.first().reset_index()
# wow
```
